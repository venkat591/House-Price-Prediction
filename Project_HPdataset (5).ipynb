## Phase 2: Data Cleaning & Manipulation

### 1) Handling Missing Data

```python
rows_with_9 = HP_df[HP_df.eq(9).any(axis=1)]                           # Extracting all rows where at least one column has value 9
rows_with_9
```

```python
HP_df =HP_df.replace(9, np.nan)                         # Replacing value 9 with NaN (treating as missing data)
```

```python
HP_df.isna().sum()                                       # Checking missing values after replacement
```

```python
HP_df.dropna(axis=0,inplace = True)                          # dropping all rows with Nan Values
```

```python
HP_df.isna().sum()
```

### 2) Removing Duplicates

```python
HP_df.info()
```

```python
HP_df.duplicated().sum()                               # Counting duplicate rows in the dataset
```

```python
HP_df=HP_df.drop_duplicates()                           # Removing duplicate rows and reassigning the DataFrame
```

```python
HP_df= HP_df.reset_index(drop=True)
```

```python
HP_df.duplicated().sum()
```

```python
HP_df.info()
```

### 3) Fix Inconsistencies in Categorical Data

```python
HP_df['Location'].unique()                                   # Checking number of unique locations
```

```python
HP_df['Location'].value_counts()                                    # Counting occurrences of each location
```

```python
HP_df['Location'] = HP_df['Location'].str.lower().str.strip()            # Converting Location column to lowercase and removing extra spaces
```

```python
HP_df['Location'].value_counts()
```

### 4) Convert Data Types


```python
HP_df.dtypes                                          # Displaying data types of all columns
```

```python
HP_df.columns
```

```python
for x in HP_df.columns:
    if x == 'Price' or x == 'Area':
        HP_df[x]=HP_df[x].astype('int32')
    elif x == 'Location':
         HP_df[x]=HP_df[x].astype('category')
    else:
         HP_df[x]=HP_df[x].astype('int8')
```
